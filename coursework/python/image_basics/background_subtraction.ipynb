```python
import imageio as io
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
```

```python
root = 'sample_pics/'
photos = os.listdir(root)
grays = []
for photo in (photos):
    this_img = io.imread(root+photo)
    this_gray = cv2.cvtColor(this_img, cv2.COLOR_BGR2GRAY)
    plt.figure()
    plt.imshow(this_img)
    grays.append(this_gray)
grays = np.array(grays)
average_grays = np.average(grays, axis=0)
plt.figure()
plt.imshow(average_grays,cmap='gray')
```

```python
for photo in (photos):
    this_img = io.imread(root+photo)
    this_gray = cv2.cvtColor(this_img, cv2.COLOR_BGR2GRAY)
    this_gray = this_gray.astype('float32')
    img_subt = this_gray - average_grays
    img_subt = np.clip(img_subt,0,255)
    plt.figure()
    plt.imshow(img_subt,cmap='gray',vmin=0,vmax=255)

    
```
